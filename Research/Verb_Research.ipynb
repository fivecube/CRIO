```python
def relevance_list_fun_2(response_eta):
    noun_verb_dict_eta = {}
    for i in range(len(response_eta['syntax']['tokens'])):
        try:
            noun_verb_dict_eta[response_eta['syntax']['tokens'][i]['lemma']] = response_eta['syntax']['tokens'][i]['part_of_speech']
        except Exception as e: print(e)
    return noun_verb_dict_eta
```

```python
def synonyms_finder(word_eta,word_pos,threshold_eta=0.5):
    import nltk
    from nltk.corpus import wordnet 
    synonyms = []

    threshold_value = threshold_eta
    verb_synset = wordnet.synset(word_eta+"."+word_pos+".01")
    
    for syn in wordnet.synsets(word_eta):
        if syn.pos() == word_pos:
            if syn.wup_similarity(verb_synset) > threshold_value:
                for l in syn.lemmas():
                    synonyms.append(l.name()) 

    return(set(synonyms))
```
