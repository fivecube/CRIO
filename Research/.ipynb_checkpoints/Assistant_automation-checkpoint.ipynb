```python
back_off = 630
while has_more:
    page = str(int(page)+1)
    print("page no",page)
    url = "https://api.stackexchange.com//2.2/tags/synonyms?key="+Auth_key+"&page="+page+"&pagesize=100&order=desc&sort=activity&site=stackoverflow"
    r = requests.get(url)
    synonym_tags = r.json()
    
    for i in range(len(synonym_tags['items'])):
        synonym_dict[synonym_tags['items'][i]['to_tag']] = synonym_tags['items'][i]['from_tag']
#         synonym_dict[synonym_tags['items'][i]['from_tag']] = synonym_tags['items'][i]['to_tag']
    has_more = synonym_tags['has_more']
    back_off-=1
```

```python
check_list = []
value_dict_eta = {}
j=0
count = 0
```

```python
refresh = len(tags_list)//20 - i
```

```python
back_off = 630
for i in range(refresh):
    j += 20
    url = api_url+"/2.2/tags/"+";".join(tags_list[j:j+20])+"/synonyms?key="+Auth_key+"&order=desc&sort=creation&site=stackoverflow"
    r = requests.get(url)
    synonym_world = r.json()
    temp_check_list = []
    if len(synonym_world['items'])!=0:
        for item in synonym_world['items']:
            if item['to_tag'] in value_dict_eta:
                value_dict_eta[str(item['to_tag'])].append(item['from_tag'])
                count+=1
                check_list.append(item['from_tag'])
                temp_check_list.append(item['from_tag'])
            else:
                item['to_tag'] = [item['from_tag']]
                count+=2
                check_list.append(item['from_tag'])
                temp_check_list.append(item['from_tag'])
                
    for item in set(tags_list[j:j+20]).difference(set(temp_check_list)):
        value_dict_eta[item] = []
        count+=1
        
    if not (back_off%90):
        print("Dictionary size is",len(value_dict_eta))
        print("Let's wait for 2 seconds")
        time.sleep(2) # this was necessary because stackoverflow throttle the ip if a lot of queries are made in one second
print("value of count is ",count)
```
